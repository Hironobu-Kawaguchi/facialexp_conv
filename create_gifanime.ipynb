# 結果をGIFアニメにする

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
file = "res/res.bmp"
images = Image.open(file)
images
```

```python
if images.mode != "RGB":
    images = images.convert("RGB") # RGBモードに変換する
```

```python
images
```

```python
imga = np.array(images)
imga.shape
```

```python
i = 5
i -= 1
imagea = imga[:,i*64:(i+1)*64,:]
image = Image.fromarray(imagea)
plt.imshow(imagea)
```

```python
num_list = list(range(12))
images = list(map(lambda num: Image.fromarray(imga[:,num*64:(num+1)*64,:]), num_list))
images[0].save('stargan_anime.gif', save_all=True, 
               append_images=images[1:], 
               duration=200, loop=0)
```
